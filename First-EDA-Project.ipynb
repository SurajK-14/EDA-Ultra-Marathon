```python
# cleanup athelete age
df2['athlete_age'] = 2020 - df2['Athlete year of birth']
# this didnt work because there are some missing values in the Athlete year of birth column
# we will need to clean this up first
```

```python
#Remove h from athelete performance
df2['Athlete performance'] = df2['Athlete performance'].str.replace('h','')
```

```python
#Drop Columns: athlete Club, Athlete Country, Athlete year of birth, Athlete age category
df2 = df2.drop(columns=['Athlete club','Athlete country','Athlete year of birth','Athlete age category'],axis=1)
```

```python
#Check for duplicate values
df2[df2.duplicated()== True]
#this shows that there are no duplicate values
```

```python
#reset index of the dataframe
df2 = df2.reset_index(drop=True)
df2.head(5)
```

```python
#fix the data types
df2.dtypes
```

```python
df2['athlete_age'] = df2['athlete_age'].astype(int)
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2.dtypes
```

```python
df2.drop(columns=['athelete_age'],axis=1,inplace=True)
```

```python
#rename columns and make everything lowercase
df2 = df2.rename(columns={'Year of event':'year',
                          'Event dates':'race_date',
                          'Event name':'race_name',
                          'Event distance/length':'race_length',
                          'Event number of finishers':'race_number_of_finishers',
                          'Athlete performance':'athlete_performance',
                          'Athlete gender':'athlete_gender',
                          'Athlete average speed':'athlete_average_speed',
                            'Athlete ID':'athlete_id',
                          })
```

```python
#Reorder columns in a new dataframe
df3 = df2[['race_date', 'race_name', 'race_length', 'race_number_of_finishers', 'athlete_id', 'athlete_gender', 'athlete_age', 'athlete_performance', 'athlete_average_speed']]
```

```python
#Difference in speed for the 50km and 50mi male to female athletes
df3.groupby(['race_length','athlete_gender'])['athlete_average_speed'].mean()
```

```python
# what age groups are the best in the 50m Races(20+ races minimum)
df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending=False).query('count > 19').head(20)
```

```python
# what age groups are the Worst in the 50m Races(20+ races minimum) (show 20)
df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending=True).query('count > 3').head(20)
```

```python
#Seasons for the data> Are athletes faster in the summer or winter?

#spring 3-5
#summer 6-8
#fall 9-11
#winter 12-2

#split between two decimals

df3['race_month']=df3['race_date'].str.split('.').str.get(1).astype(int)
df3['race_season']= df3['race_month'].apply(lambda x: 'spring' if x in [3,4,5] else 'summer' if x in [6,7,8] else 'fall' if x in [9,10,11] else 'winter' if x in [12,1,2] else 'unknown')
df3.head(30)
```

```python
df3.groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending=False)
```

```python
df3.query('race_length == "50mi"').groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending=False)
```
